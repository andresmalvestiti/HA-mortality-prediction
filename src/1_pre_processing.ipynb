Procedemos a ver cuales son las columnas tipo `object` y hacemos una limpieza de los datos erroneos, nulos o vacios. 

```python
# Obtenemos las columnas de tipo object
object_columns = df.select_dtypes(include=['object']).columns

df[object_columns].info()
```

Vemos que tipo de datos unicos tenemos en cada columna

```python
for column in object_columns:
  print(f"{column}: {df[column].unique()}")
```

Observamos tipos de lo mas variados: numeros, strings, dates, nan.

Vemos que hay valores string vacios, strings que representan una expresion numerica `<0,01`, guiones, texto aclaratorio, etc.
Tenemos que limpiar todos los simbolos y espacios en blanco. Tambien transformar los textos en valores que tengan sentido para esa feature. 

```python
# Aplicamos la función solo a las columnas definidas en object_columns
df[object_columns] = df[object_columns].applymap(clean_value)
```

```python
for column in object_columns:
  print(f"{column}: {df[column].unique()}")
```

Luego de limpiar los valores no deseados, pasamos todas las columnas a float (tipo de dato numérico que admite NaN) asi podemos empezar la imputación de los valores faltantes en el proximo notebook.

```python
# Convertir las columnas a float para admitir NaN
df[object_columns] = df[object_columns].apply(pd.to_numeric, errors='coerce')
```
