# Data split
En esta sección vamos a hacer el split en train y test para nuestros futuros experimentos. El resultado van a ser 4 archivos. Dos archivos para train y test con X e y.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/imputed_data.csv').drop(columns=["Unnamed: 0"])
```

```python
df.head()
```

```python
target = 'MUERTE HOSP'
```

```python
X = df.drop(columns=target)
y = df[[target]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0, stratify=y)
```

```python
# Guardar las características de entrenamiento y prueba
X_train.to_csv("../data/sca_X_train.csv", index=False)
X_test.to_csv("../data/sca_X_test.csv", index=False)

# Guardar las etiquetas (target) de entrenamiento y prueba
y_train.to_csv("../data/sca_y_train.csv", index=False)
y_test.to_csv("../data/sca_y_test.csv", index=False)
```
