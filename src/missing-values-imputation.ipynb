La imputación de los datos la vamos a hacer en dos partes. Para aquellas features que tienen valores faltantes mayores o iguales al 2% del total, imputamos mediante la técnica MICE (Multiple Imputation by Chained Equations). Para las features con valores faltantes menores a 2%, hacemos una imputación aleatoria ya que no afecta la distribución de la feature en particular.

```python
# Lista de columnas para imputar mediante MICE
mice_imputation_columns = ['PESO Kg', 'ALTURA cm', 'GLUCEMIA INGR', 'GB']
```

### Imputacion Univariada - Imputación por muestreo aleatorio

```python
# Creamos una copia del dataframe
df_copy = df.copy()
```

```python
nan_columns = get_columns_with_nan(df_copy)
```

```python
# Obtenemos las columnas que contienen NaN menor a 2%
random_imputation_columns = [col for col in nan_columns if col not in mice_imputation_columns]
```

```python
for column in random_imputation_columns:
    # Obtenemos los valores no NaN de la columna
    no_nan_values = df_copy[column].dropna()
    # Imputamos los valores NaN con muestreo aleatorio de los valores no NaN
    df_copy[column] = df_copy[column].apply(
        lambda x: np.random.choice(no_nan_values) if np.isnan(x) else x
    )
```

```python
print_column_with_nan(df_copy)
```

### MICE (Multiple Imputation by Chained Equations)

- Calcula valores faltantes para cada columna como una regresión lineal de las restantes (si bien también se pueden usar otros modelos). Como utiliza el resto de las columnas, es un método **multivariado**.
- Es iterativo porque repite este proceso hasta que la diferencia entre las dos últimas iteraciones sea cercana a cero.

Búsqueda de mejores parámetros por Grid Search.

```python
X = df_copy.drop([TARGET], axis=1)
y = df_copy[TARGET]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.experimental import enable_iterative_imputer
from sklearn.model_selection import GridSearchCV, RepeatedStratifiedKFold
from sklearn.impute import IterativeImputer

# Definir el pipeline con imputación iterativa y regresión logística
pipeline = Pipeline(steps=[
    ('imputer', IterativeImputer(max_iter=10, random_state=0)),
    ('lr', LogisticRegression(max_iter=1000))
])

# Definir los parámetros para la búsqueda en cuadrícula
param_grid = {
    "imputer__n_nearest_features": [None] + list(range(1, X.shape[1] + 1))
}

# Configurar la validación cruzada repetida
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=5, random_state=42)

# Configurar la búsqueda en cuadrícula
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=cv, scoring="f1", n_jobs=-1)

# Ajustar el modelo a los datos de entrenamiento
grid_search.fit(X, y)

# Imprimir los mejores parámetros y el mejor puntaje F1
print("Mejores parámetros encontrados:")
print(grid_search.best_params_)
print("\nMejor puntaje F1:")
print(grid_search.best_score_)
```

```python
best_mice_imputer = grid_search.best_estimator_['imputer']

# Imputar solo las columnas de mice_imputation_columns
X_imputed = best_mice_imputer.transform(X)
df_imputed = df_copy.copy()
df_imputed[mice_imputation_columns] = X_imputed[:, [df_copy.columns.get_loc(col) for col in mice_imputation_columns]]
```
